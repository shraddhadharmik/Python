```python
se = pd.Series([1,2,3,5,6])
print(se)
```

```python
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/pscl/ca2006.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#% of data missing from dataset

missing_data = df.isnull().sum()
missing_data.sum()/(df.shape[0] * df.shape[1]) * 100
```

```python
#drop row where values are null

df.dropna()
```

```python
#df.drop(axis = 1, how='any')
#df.drop(axis = 1, how ='all')

#atleast 1 not nan value
#df.drop(thresh = 1)
```
