```python
data = pd.read_csv('https://parvathy47.github.io/Parvathy47.github.io/Pak_Suicide_Bombings.csv')
```

```python
province = data['Province'].unique()
province
```

```python
data['Province'] =  data['Province'].str.lower()
data['Province'] =  data['Province'].str.strip()
```

```python
type(data['Province'].str)
```

```python
import fuzzywuzzy
from fuzzywuzzy import process
```

```python
matches = fuzzywuzzy.process.extract('atfa', province, scorer = fuzzywuzzy.fuzz.token_sort_ratio)
matches
```

```python
fuzzywuzzy.process.extract('balochistan', ['baluchistan'], scorer = fuzzywuzzy.fuzz.token_sort_ratio, limit=10)
```

```python
#Function to replace strings based on ratio match

def replace_macthes(df, column, string_to_match, min_ratio = 90) :
    strings = df[column].unique()
    matches = fuzzywuzzy.process.extract(string_to_match, strings, scorer = fuzzywuzzy.fuzz.token_sort_ratio, limit=len(strings))
    close_matches = [match[0] for match in matches if match[1]>=min_ratio]
    match_rows = df[column].isin(close_matches)
    df.loc[match_rows, column] = string_to_match
    print('All done!')
```

```python
replace_macthes(data, 'Province', 'balochistan')
```
