```python
import pandas as pd
import numpy as np
```

```python
#1. Read data from https://vincentarelbundock.github.io/Rdatasets/csv/Stat2Data/NCbirths.csv

df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/Stat2Data/NCbirths.csv')
```

```python
#2. Print the shape of the data

df.shape
```

```python
#3. Print columns that contain null values with the number of null values in each.

for col in df.columns :
    if df[col].isnull().sum() > 0 :
        print("{} : {} ".format(col,df[col].isnull().sum()))
```

```python
#4. Print the rows that contain maximum number of NaN values


index = -1
count = -1
for i in range(df.shape[0]) :
    if df.loc[i].isnull().sum() > count :
        count = df.loc[i].isnull().sum()
        index = i
        
print('Row with maximum null values : \n', df.loc[index])

#Alternate
df[df.isnull().sum(axis=1) == max(df.isnull().sum(axis=1))]
```

```python
#5. Print the count of male and female babies

print('Male Count : ', df[df['Sex'] == 1].shape[0])
print('Female Count : ', df[df['Sex'] == 2].shape[0])
```

```python
#6. Print the percentage of missing values in each column

for col in df.columns :
    print("{} : {}% ".format(col, np.round((df[col].isnull().sum()/df.shape[0])*100, 2)))
```

```python
#7. How many columns are dropped if we decide to eliminate columns that contain more than 1% of null values?

count = 0
for col in df.columns :
    if np.round((df[col].isnull().sum()/df.shape[0])*100, 2) > 1 :
        count += 1
        
print('Columns greater than 1% that could be dropped : ', count)
```

```python
#8. How many rows are dropped if we decide to eliminate rows that contain more than 5% of null values?

count = 0
for i in range(df.shape[0]) :
    if (df.loc[i].isnull().sum() / df.shape[1])*100 > 5 :
        count += 1
        print('----Row----', df.loc[i])
print('Count : ', count)
```

```python
#9. Print the dataframe by filling NaN values with the mean of the column

for col in df.columns :
    if df[col].isnull().sum()/df.shape[0] > 0 :
        df[col].fillna(df[col].mean())
```

```python
#10. Print the dataframe using the back fill method

df.fillna(method='bfill')
```
