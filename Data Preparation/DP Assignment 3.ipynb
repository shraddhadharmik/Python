```python
data1 = {'dept_id':[10,20,30,40,50], 'dept_name':['Administration','Marketing','Human Resources','IT','Public Relations'],
        'loc_id' : [1700,1800,2400,1700,2400]}

df1 = pd.DataFrame(data1)
df1
```

```python
pd.merge(df1, df2, on = 'loc_id')
```
