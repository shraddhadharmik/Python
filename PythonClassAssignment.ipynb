# Arguments passed using * and **
 - \* will take the arguments as it is
 - ** will take values that we have sent with a variable name and is taken as dictionary

```python
def code(*args, **kwargs) :
    for i in args :
        print(i)
    for k,v in kwargs.items() :
        print(k,v)
        
code(5, 'Hello', (1,0,2), a='l', l=[5,3,2])
```
